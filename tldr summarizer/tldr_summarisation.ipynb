````python
import openai
from bs4 import BeautifulSoup
import requests
````

````python
def extract_text_from_webpage(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    text = ' '.join(soup.stripped_strings)
    return text


# url = input("Enter the URL of the webpage: ")
# webpage_text = extract_text_from_webpage(url)
# webpage_text
````

````python
def get_webpage_content(url):
    try:
        response = requests.get(url)
        return response.text
    except requests.exceptions.RequestException:
        print("Error: Failed to fetch webpage.")
        return None

# url = input("Enter the URL of the webpage: ")
# webpage_text = get_webpage_content(url)
# webpage_text
````

````python

# Function to generate a summary using ChatGPT
def generate_summary(prompt):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a helpful assistant that summarizes webpages."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=100
    )
    return response.choices[0].message.content
````

````python
# Function to answer questions using ChatGPT
delimiter = "```"
template = "you only answer the questions that can be found in the information given between three backticks(```). Do not answer question outside the contex."

def answer_question(question, context):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a knowledgeable assistant that answers questions, "+{template}},
            {"role": "assistant", "content": delimiter+context+delimiter},
            {"role": "user", "content": question}
        ],
        max_tokens=100
    )
    return response.choices[0].message.content
````

````python
# Main program

# Read the webpage content
url = input("Enter the URL of the webpage: ")
webpage_content = extract_text_from_webpage(url)

# Summarize the webpage content
summary_text = generate_summary(webpage_content)
print(f"Webpage summary:\n{summary_text}\n")

# Ask user questions and get answers based on the summary
while True:
    user_question = input("Ask a question (or 'exit' to quit): ")
    if user_question.lower() == 'exit':
        break
    answer = answer_question(summary_text, user_question)
    print(f"Answer: {answer}\n")
````
